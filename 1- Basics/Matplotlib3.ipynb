```python
year=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.895,0.91,0.919,0.926,0.929,0.931]
yield_orange=[0.962,0.941,0.923,0.928,0.956,0.962]
plt.plot(year,yield_apples,marker='o')
plt.plot(year,yield_orange,marker='*')
plt.xlabel('Years')
plt.ylabel('Yield')
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset('tips')
```

```python
data
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
#countplot
sns.countplot(data=data,x='smoker',palette=['red','green'])
```

```python
#Distribution Plot
sns.distplot(data.total_bill)
```

```python
sns.distplot(data.total_bill, hist=False)
```

```python
sns.distplot(data.total_bill, kde=False)
```

```python
sns.scatterplot(x='total_bill',y='tip', data=data, hue='sex', size='smoker', palette=['green', 'red'])
```

```python
# Joint Plot
sns.jointplot(x='total_bill',y='tip', data=data)
```

```python
#box plot
sns.boxplot(data.total_bill, orient='h')
```

```python
#pair plor- 
sns.pairplot(data)
```

```python
#corr() # to find the correlation  between variables
data.corr(numeric_only=True)
```

```python
# heat map
hm=data.corr(numeric_only=True)   #data.corr()
```

```python
sns.heatmap(hm, annot=True)
```
